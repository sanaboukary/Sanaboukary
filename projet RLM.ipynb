  <h3 style="color:#2c3e50; margin-top:0;">üìå Quand privil√©gier l‚Äôimputation par la moyenne ?</h3>

  <ul style="font-size:15px; color:#333; line-height:1.7; padding-left:20px;">
    <li><strong>Distribution sym√©trique</strong> : La variable ne pr√©sente pas de forte asym√©trie ni de valeurs extr√™mes.</li>
    <li><strong>Donn√©es num√©riques continues</strong> : Type <code>float</code> ou <code>int</code>, excluant les variables ordinales ou cat√©gorielles.</li>
    <li><strong>Taux de valeurs manquantes faible</strong> : Moins de 5 √† 10 % des donn√©es manquantes pour limiter le biais.</li>
    <li><strong>Phase exploratoire</strong> : Pour visualiser rapidement les donn√©es sans affecter l‚Äôanalyse finale.</li>
    <li><strong>Mod√®les robustes</strong> : Si les algorithmes utilis√©s sont peu sensibles aux distributions (ex. : arbres de d√©cision).</li>
  </ul>

  <p style="font-size:15px; color:#2c3e50; line-height:1.6;">
    ‚ö†Ô∏è <strong>√Ä √©viter</strong> en pr√©sence de <em>valeurs aberrantes</em> ou de <em>distributions tr√®s asym√©triques</em>, car la moyenne peut fausser la repr√©sentation centrale.
  </p>
</div>


<div style="font-family:'Segoe UI', sans-serif; background-color:#fffef6; padding:20px; border-left:6px solid #f39c12; border-radius:10px; margin:30px 0;">

  <h3 style="color:#c87f0a; margin-top:0;">üìç Justification du choix : Imputation par la m√©diane</h3>

  <p style="font-size:15px; color:#333; line-height:1.6;">
    Pour le traitement des valeurs manquantes de la variable <strong>Courier_Experience_yrs</strong>, nous avons opt√© pour une <strong>imputation par la m√©diane</strong>, plut√¥t que par la moyenne.
    Ce choix est motiv√© par plusieurs raisons statistiques :
  </p>

  <ul style="font-size:15px; color:#444; line-height:1.7; padding-left:20px;">
    <li><strong>Robustesse face aux valeurs extr√™mes</strong> : La m√©diane n‚Äôest pas influenc√©e par des valeurs aberrantes, contrairement √† la moyenne.</li>
    <li><strong>Distribution l√©g√®rement asym√©trique</strong> : Une inspection graphique et statistique r√©v√®le que les ann√©es d‚Äôexp√©rience ne suivent pas une distribution parfaitement sym√©trique.</li>
    <li><strong>Pr√©servation de la centralit√© r√©elle</strong> : La m√©diane refl√®te mieux la tendance centrale lorsque les donn√©es sont dispers√©es ou h√©t√©rog√®nes.</li>
    <li><strong>√âcart-type √©lev√©</strong> : La forte dispersion observ√©e sugg√®re que la moyenne pourrait sur√©valuer ou sous-√©valuer les donn√©es manquantes.</li>
  </ul>

  <p style="font-size:15px; color:#2c3e50; line-height:1.6;">
    üîé Ainsi, l‚Äôutilisation de la m√©diane permet de <strong>minimiser les biais</strong> et de <strong>pr√©server la repr√©sentativit√©</strong> de la variable dans les analyses ult√©rieures, notamment dans les mod√®les de r√©gression.
  </p>
</div>


<div style="font-family:'Segoe UI', sans-serif; background-color:#fdfcf7; padding:20px; border-left:6px solid #f39c12; border-radius:10px; margin:30px 0;">

  <h3 style="color:#c87f0a; margin-top:0;">üìä Justification par le code Python</h3>

  <p style="font-size:15px; color:#333; line-height:1.6;">
    Afin de d√©terminer la strat√©gie d‚Äôimputation la plus adapt√©e pour la variable <strong>Courier_Experience_yrs</strong>, nous avons compar√© la diff√©rence entre les deux principales mesures de tendance centrale (<code>mean</code> et <code>median</code>) et l‚Äô√©cart-type (<code>std</code>) √† travers les calculs suivants :
  </p>

  <pre style="background-color:#f5f5f5; padding:10px; border-radius:6px; font-size:14px; overflow-x:auto;">
mean = livraison["Courier_Experience_yrs"].mean()
std = livraison["Courier_Experience_yrs"].std()
variation_mean = mean - std  # ‚û§ R√©sultat : 1.66

median = livraison["Courier_Experience_yrs"].median()
variation_median = median - std  # ‚û§ R√©sultat : 2.09
  </pre>

  <p style="font-size:15px; color:#2c3e50; line-height:1.6;">
    üìå <strong>Interpr√©tation :</strong> La valeur <code>median - std</code> (‚âà 2.09) est sensiblement plus √©lev√©e que <code>mean - std</code> (‚âà 1.66), 
    ce qui indique que la <strong>m√©diane est moins influenc√©e par la dispersion</strong> dans les donn√©es.
  </p>

  <p style="font-size:15px; color:#2c3e50; line-height:1.6;">
    ‚öñÔ∏è Par cons√©quent, la <strong>m√©diane est plus repr√©sentative de la "tendance r√©elle"</strong> des ann√©es d‚Äôexp√©rience dans ce contexte. Ce raisonnement statist


```python
median_val = livraison["Courier_Experience_yrs"].median()
livraison.loc[livraison["Courier_Experience_yrs"].isna(), "Courier_Experience_yrs"] = median_val
livraison["Courier_Experience_yrs"].isnull().sum()
```

  <h3 style="color:#2c3e50; margin-top:0;">üå§Ô∏è Imputation par le mode pour la variable <code>Weather</code></h3>

  <p style="font-size:15px; color:#333; line-height:1.6;">
    L‚Äôanalyse de la variable qualitative <strong>Weather</strong> r√©v√®le les √©l√©ments suivants :
  </p>

  <ul style="font-size:15px; color:#444; line-height:1.7;">

    <li><strong>Distribution des modalit√©s :</strong> La modalit√© <code>Clear</code> est largement dominante avec <strong>48.5 %</strong> des occurrences, devant <code>Rainy</code> (21.3 %), <code>Foggy</code> (10.6 %), <code>Snowy</code> (10.0 %) et <code>Windy</code> (9.6 %).</li>
  </ul>

  <p style="font-size:15px; color:#2c3e50; line-height:1.6;">
    ‚úÖ √âtant donn√© la <strong>faible dispersion</strong> des modalit√©s et la <strong>domination de la modalit√© "Clear"</strong>, il est statistiquement pertinent d‚Äôeffectuer une <strong>imputation par le mode</strong> :

  <pre style="background-color:#f4f4f4; padding:10px; border-radius:6px; font-size:14px; overflow-x:auto;">
livraison["Weather"].fillna(livraison["Weather"].mode()[0], inplace=True)

  <h3 style="color:#7d3c98; margin-top:10px;">üìà Pourquoi avoir choisi la m√©thode IQR pour d√©tecter les valeurs aberrantes ?</h3>

  <p style="font-size:15px; color:#333; line-height:1.6;">
    Dans le cadre de notre projet d‚Äôanalyse √©conom√©trique du <strong>d√©lai de livraison</strong>, la m√©thode <strong>IQR (Interquartile Range)</strong> a √©t√© retenue pour identifier les valeurs extr√™mes. Ce choix repose sur plusieurs justifications solides :
  </p>

  <ul style="font-size:15px; color:#444; line-height:1.7;">
    <li><strong>Adapt√©e aux variables continues</strong> : Nos indicateurs comme <code>Delivery_Time_min</code> ou <code>Preparation_Time_min</code> sont des dur√©es susceptibles de contenir des retards ou anomalies ponctuelles.</li>
    <li><strong>Robuste face aux distributions asym√©triques</strong> : Contrairement au z-score, l‚ÄôIQR ne d√©pend pas de la moyenne ou de l‚Äô√©cart-type, ce qui est id√©al pour des donn√©es non normales.</li>
    <li><strong>Simplicit√© et clart√© d‚Äôinterpr√©tation</strong> : Elle identifie les valeurs situ√©es en dehors de l‚Äôintervalle [Q1 - 1.5√óIQR ; Q3 + 1.5√óIQR], ce qui facilite la lecture des r√©sultats.</li>
    <li><strong>Pr√©serve l'int√©grit√© des donn√©es</strong> : Elle permet d‚Äôisoler les outliers pour analyse sans automatiquement les exclure.</li>
  </ul>

  <p style="font-size:15px; color:#2c3e50; line-height:1.6;">
    ‚ö†Ô∏è Bien que ce seuil (1.5√óIQR) soit arbitraire, il peut √™tre adapt√© selon les besoins m√©tier (ex. : 2√óIQR pour tol√©rer plus de variabilit√©).
  </p>

  <p style="font-size:15px; color:#2c3e50;">
    ‚úÖ En r√©sum√©, la m√©thode IQR offre une solution <strong>fiable, robuste et adapt√©e</strong> √† la d√©tection d‚Äôobservations atypiques dans notre contexte, sans pert


```python

# √âtape 2 : Fonction de d√©tection des valeurs aberrantes
def detect_outliers_iqr(df, seuil=1.5):
    outliers_info = []

    # On s√©lectionne uniquement les colonnes num√©riques
    num_cols = df.select_dtypes(include=['float64', 'int64']).columns

    for col in num_cols:
        Q1 = df[col].quantile(0.25)
        Q3 = df[col].quantile(0.75)
        IQR = Q3 - Q1
        borne_inf = Q1 - seuil * IQR
        borne_sup = Q3 + seuil * IQR

        # Identification des outliers
        outliers = df[(df[col] < borne_inf) | (df[col] > borne_sup)]
        nb_outliers = outliers.shape[0]
        pourcentage = (nb_outliers / df.shape[0]) * 100

        outliers_info.append({
            "Variable": col,
            "Nb_outliers": nb_outliers,
            "Pourcentage (%)": round(pourcentage, 2)
        })

    return pd.DataFrame(outliers_info)

# √âtape 3 : Application de la fonction sur le DataFrame
outliers_df = detect_outliers_iqr(livraison)

# √âtape 4 : Affichage du tableau r√©sultat
print("üîç Tableau des valeurs aberrantes (m√©thode IQR) :")
display(outliers_df)
```

```python
from scipy.stats.mstats import winsorize
import seaborn as sns
import matplotlib.pyplot as plt

# üßÆ S√©lection des variables quantitatives (hors identifiant)
quantitative_vars = livraison.select_dtypes(include=['float64', 'int64']).columns.tolist()
quantitative_vars = [col for col in quantitative_vars if col != "Order_ID"]

# üîÅ Cr√©ation d'une copie pour comparaison avant traitement
livraison_before = livraison.copy()

# üß™ Application de la winsorisation (1 % des extr√™mes √† chaque bout)
for col in quantitative_vars:
    livraison[col] = winsorize(livraison[col], limits=(0.01, 0.01))

# üìä Comparaison graphique AVANT vs APR√àS winsorisation
plt.figure(figsize=(16, 10))

# üîπ Boxplot avant winsorisation
plt.subplot(1, 2, 1)
sns.boxplot(data=livraison_before[quantitative_vars], orient='h', palette="coolwarm")
plt.title("üì¶ Avant winsorisation", fontsize=14)
plt.grid(axis='x', linestyle='--', alpha=0.5)

# üîπ Boxplot apr√®s winsorisation
plt.subplot(1, 2, 2)
sns.boxplot(data=livraison[quantitative_vars], orient='h', palette="coolwarm")
plt.title("üì¶ Apr√®s winsorisation", fontsize=14)
plt.grid(axis='x', linestyle='--', alpha=0.5)

plt.tight_layout()
plt.show()

# ‚úÖ V√©rification finale
print("‚úÖ Donn√©es apr√®s winsorisation (aper√ßu) :")
display(livraison[quantitative_vars].head())
```

  <h3 style="color:#1f618d; margin-top:0;">1. üîπ Introduction au mod√®le</h3>

  <p style="font-size:15px; color:#2c3e50; line-height:1.7;">
    üéØ <strong>Objectif :</strong> Expliquer les variations du <strong>d√©lai de livraison</strong> √† partir de plusieurs variables explicatives influentes, tant quantitatives que qualitatives.
  </p>

  <p style="font-size:15px; color:#2c3e50; line-height:1.7;">
    üí° <strong>M√©thode utilis√©e :</strong> R√©gression lin√©aire multiple (RLM)
  </p>

  <p style="font-size:15px; color:#2c3e50; line-height:1.7;">
    üîç <strong>Justification :</strong> Cette m√©thode est particuli√®rement adapt√©e pour mod√©liser une variable quantitative continue 
    (<em>le d√©lai de livraison</em>) en fonction de plusieurs facteurs pr√©dictifs simultan√©s, qu‚Äôils soient <strong>num√©riques</strong> 
    (ex. : distance, temps de pr√©paration) ou <strong>cat√©goriels</strong> (ex. : m√©t√©o, type de v√©hicule, moment de la journ√©e...).
  </p>

</div>


<section style="background-color:#f9f9f9; padding: 1.2em; border-left: 5px solid #2ecc71; border-radius: 8px; font-family: 'Segoe UI', sans-serif;">
  <h2 style="color:#2c3e50;">
    ‚úÖ 2. Choix de la m√©thode d‚Äôestimation
  </h2>
  <p style="margin-top: 0.5em; font-size: 1.1em;">
    üìä On utilise la m√©thode des <strong>Moindres Carr√©s Ordinaires (MCO)</strong> :
  </p>
  <blockquote style="margin: 0.8em 0; padding: 0.8em 1em; border-left: 4px solid #ccc; background-color: #fff;">
    Elle permet d‚Äôestimer les coefficients en minimisant la somme des carr√©s des r√©sidus.
  </blockquote>
</section>


<div style="font-family:'Segoe UI', sans-serif; background-color:#fdfefe; padding:20px; border-left:6px solid #117864; border-radius:8px; margin:30px 0;">

  <h3 style="color:#117864;">2. üß™ Pr√©paration des donn√© et estimation du modeles</h3>

  <h4 style="margin-top:20px;">a. S√©lection des variables explicatives pertinentes</h4>
  <ul style="font-size:15px; color:#2c3e50; line-height:1.7;">
    <li>
      üîπ <strong>Variables quantitatives :</strong> 
      <code>Distance_km</code>, <code>Preparation_Time_min</code>, <code>Courier_Experience_yrs</code>
    </li>
    <li>
      üîπ <strong>Variables qualitatives (√† encoder) :</strong> 
      <code>Weather</code>, <code>Traffic_Level</code>, <code>Time_of_Day</code>, <code>Vehicle_Type</code>
    </li>
  </ul>

  <h4 style="margin-top:20px;">b. Encodage des variables qualitatives</h4>
  <p style="font-size:15px; color:#2c3e50;">
    üìå M√©thode : <code>pd.get_dummies(..., drop_first=True)</code> afin d‚Äô√©viter la multicolin√©arit√©.
  </p>

  <h4 style="margin-top:20px;">c. D√©finition des matrices</h4>

    X = livraison_encoded.drop(columns="Delivery_Time_min")  <span style="color:#7f8c8d;"># variables explicatives</span><br>
    y = livraison_encoded["Delivery_Time_min"]                <span style="color:#7f8c8d;"># variable cible</span>
  </div>

```python
import statsmodels.api as sm

# ‚úÖ Conversion explicite en float pour √©viter les erreurs avec les bool√©ens
X_const = sm.add_constant(X).astype(float)
y = y.astype(float)

# ‚úÖ Estimation du mod√®le
model = sm.OLS(y, X_const).fit()

# ‚úÖ R√©sum√© du mod√®le
print(model.summary())
```

    Dans cette section, nous cherchons √† d√©terminer si la <strong>m√©t√©o</strong> a un effet significatif sur le <strong>temps de livraison</strong> des commandes.
    Pour cela, nous utilisons une <strong>analyse de variance (ANOVA)</strong> √† un facteur, o√π :
  </p>
  <ul style="margin-left: 20px; font-size: 15px;">
    <li><strong>Variable d√©pendante :</strong> <code>Delivery_Time_min</code> (quantitative continue)</li>
    <li><strong>Facteur explicatif :</strong> <code>Weather</code> (qualitative √† plusieurs modalit√©s : Sunny, Rainy, Foggy, Snowy, Windy...)</li>
  </ul>
  <p style="font-size: 15px;">
    Le but est de tester l'hypoth√®se :
  </p>
  <blockquote style="background-color: #e8f1ff; padding: 10px; border-left: 4px solid #007ACC;">
    <strong>H<sub>0</sub> :</strong> Les moyennes des d√©lais de livraison sont √©gales pour toutes les conditions m√©t√©o<br>
    <strong>H<sub>1</sub> :</strong> Au moins une condition m√©t√©o induit un d√©lai significativement diff√©rent
  </blockquote>
  <p style="font-size: 15px;">
    En cas de rejet de l'hypoth√®se nulle, nous proc√©derons √† des <strong>tests post hoc</strong> (Tukey, Scheff√©...) pour identifier les diff√©rences sp√©cifiques entre les groupes.
  </p>
</div>


<div style="border: 2px solid #4caf50; border-radius: 8px; padding: 16px; background-color: #f9fff9; font-family: Arial, sans-serif;">
  <h2 style="color: #2e7d32;">üìå 2. Pr√©paration des donn√©es</h2>

  <h3 style="margin-top: 10px;">a. V√©rification des variables concern√©es</h3>
  <ul>
    <li>‚úÖ <strong>Variable d√©pendante :</strong> <code>Delivery_Time_min</code> <em>(quantitative)</em></li>
    <li>‚úÖ <strong>Variable explicative :</strong> <code>Weather</code> <em>(qualitative √† plusieurs modalit√©s)</em></li>
  </ul>

  <h3 style="margin-top: 10px;">b. Statistiques descriptives par groupe</h3>
  <ul>
    <li>üìä Moyenne, m√©diane, √©cart-type du d√©lai pour chaque type de m√©t√©o</li>
    <li>üìâ Visualisation : Boxplot ou Violin Plot group√© par <code>Weather</code></li>

</div>

<div style="border: 2px solid #4CAF50; border-radius: 8px; padding: 18px; background-color: #f9fff9;">
  <h2 style="color: #2e7d32;">üîé Analyse Pr√©paratoire ‚Äì Livraison vs M√©t√©o</h2>

  <p>
    Avant d'appliquer un mod√®le ANOVA, les donn√©es ont √©t√© soigneusement nettoy√©es et pr√©par√©es :
  </p>
  
  <ul>
    <li>‚úÖ <strong>Suppression des valeurs manquantes</strong> sur les variables <code>Weather</code> et <code>Delivery_Time_min</code>.</li>
    <li>‚úÖ <strong>Winsorisation</strong> du d√©lai de livraison (<code>Delivery_Time_min</code>) afin de limiter l‚Äôinfluence des valeurs extr√™mes (1% aux deux extr√©mit√©s).</li>
    <li>üìä <strong>Statistiques descriptives</strong> calcul√©es pour chaque condition m√©t√©o : moyenne, m√©diane, √©cart-type et effectif.</li>
  </ul>

  <p>
    La <strong>visualisation par boxplot</strong> montre des diff√©rences notables dans la distribution du d√©lai selon les conditions m√©t√©o. Par exemple :
  </p>

  <ul style="margin-left: 15px;">
    <li>üåßÔ∏è <strong>Temps pluvieux</strong> ou <strong>neigeux</strong> ‚Üí tendance √† allonger significativement le d√©lai de livraison.</li>
    <li>‚òÄÔ∏è <strong>Temps clair ou venteux</strong> ‚Üí d√©lai souvent plus faible et plus concentr√©.</li>
  </ul>

  <p style="margin-top: 12px;">
    Cette premi√®re √©tape confirme l‚Äôint√©r√™t d‚Äôun <strong>test d‚Äô√©galit√© des moyennes</strong> (ANOVA) pour valider statistiquement l‚Äôeffet de la m√©t√©o sur les d√©lais.
  </p>
</div>


<div style="border-left: 5px solid #1976d2; background-color: #f4f9ff; padding: 16px; font-family: Arial, sans-serif;">
  <h2 style="color: #0d47a1;">üìò 3. Hypoth√®ses de l‚ÄôANOVA</h2>

  <ul style="line-height: 1.7;">
    <li>
      <strong>H<sub>0</sub> (hypoth√®se nulle)</strong> :<br>
      Les moyennes de <code>Delivery_Time_min</code> sont identiques entre les groupes de <code>Weather</code>.
    </li>
    <li>
      <strong>H<sub>1</sub> (hypoth√®se alternative)</strong> :<br>
      Au moins une moyenne de <code>Delivery_Time_min</code> diff√®re significativement.
    </li>
  </ul>
</div>


<div style="border-left: 5px solid #1565c0; background-color: #f2f7fd; padding: 16px; font-family: Arial, sans-serif;">
  <h2 style="color: #0d47a1;">üîç 4. Conditions d‚Äôapplication de l‚ÄôANOVA</h2>

  <ol style="line-height: 1.7;">
    <li>
      <strong>Normalit√© des r√©sidus</strong> :<br>
      ‚úîÔ∏è V√©rifi√©e par un test de <em>Shapiro-Wilk</em> ou par l‚Äôanalyse des histogrammes des r√©sidus par groupe.
    </li>
    <li>
      <strong>Homog√©n√©it√© des variances</strong> :<br>
      ‚úîÔ∏è √Ä tester via le <em>test de Levene</em> ou de <em>Bartlett</em>.
    </li>
    <li>
      <strong>Ind√©pendance des observations</strong> :<br>
      ‚úîÔ∏è D√©pend du protocole exp√©rimental, mais souvent suppos√©e si les donn√©es sont bien √©chantillonn√©es al√©atoirement.
    </li>
  </ol>
</div>


```python
import pandas as pd
import numpy as np
from scipy.stats import shapiro, levene
from scipy.stats.mstats import winsorize
from statsmodels.formula.api import ols
from statsmodels.stats.anova import anova_lm

# Conversion de Weather en variable cat√©gorielle
livraison["Weather"] = livraison["Weather"].astype("category")

# Test de normalit√© (Shapiro-Wilk) pour chaque modalit√©
shapiro_results = livraison.groupby("Weather")["Delivery_Time_min"].apply(lambda x: shapiro(x)[1]).reset_index()
shapiro_results.columns = ["Weather", "Shapiro_p_value"]

# Test de Levene pour homog√©n√©it√© des variances
groups = [g["Delivery_Time_min"].values for _, g in livraison.groupby("Weather")]
levene_stat, levene_p = levene(*groups)

# Test ANOVA
model = ols("Delivery_Time_min ~ C(Weather)", data=livraison).fit()
anova_table = anova_lm(model)

# R√©sum√©
summary_livraison = pd.DataFrame({
    "Test": ["Shapiro-Wilk (min)", "Shapiro-Wilk (max)", "Levene", "ANOVA"],
    "Valeur": [
        shapiro_results["Shapiro_p_value"].min(),
        shapiro_results["Shapiro_p_value"].max(),
        round(levene_p, 4),
        round(anova_table["PR(>F)"][0], 4)
    ],
    "Interpr√©tation attendue": [
        "> 0.05 pour normalit√©",
        "> 0.05 pour normalit√©",
        "> 0.05 pour homog√©n√©it√©",
        "< 0.05 pour effet significatif"
    ]
})

# Affichage des r√©sultats
display(shapiro_results)
display(summary_livraison)
```

<div style="border: 2px solid #007bff; border-radius: 8px; padding: 15px; background-color: #f5faff;">
  <h3 style="color: #0056b3;">‚úÖ R√©sum√© des V√©rifications Pr√©liminaires √† l‚ÄôANOVA</h3>
  <ul>
    <li><strong>Homog√©n√©it√© des variances</strong> : Confirm√©e (Test de Levene, p = 0.76).</li>
    <li><strong>Normalit√© des r√©sidus</strong> : Non respect√©e pour certains groupes (p < 0.05), mais l‚ÄôANOVA reste robuste avec un grand effectif.</li>
    <li><strong>Effet significatif</strong> : L‚ÄôANOVA indique une influence r√©elle de la m√©t√©o (<code>Weather</code>) sur le d√©lai de livraison (<code>Delivery_Time_min</code>), p < 0.0001.</li>
  </ul>
  <div style="background-color: #e6f2ff; padding: 10px; margin-top: 10px; border-left: 4px solid #007bff;">
    <strong>üëâ Recommandation :</strong> Lancer une <em>analyse post hoc</em> (ex. Tukey HSD) pour d√©tecter les diff√©rences pr√©cises entre types de m√©t√©o.
  </div>
</div>


```python
# üì¶ Imports
import pandas as pd
import numpy as np
from scipy.stats.mstats import winsorize
from sklearn.impute import SimpleImputer
from statsmodels.formula.api import ols
from statsmodels.stats.multicomp import pairwise_tukeyhsd
import statsmodels.api as sm
import matplotlib.pyplot as plt
import seaborn as sns


# 5Ô∏è‚É£ Filtrage sur la cible
q1, q3 = livraison["Delivery_Time_min"].quantile([0.01, 0.99])
df = livraison[(livraison["Delivery_Time_min"] >= q1) & (livraison["Delivery_Time_min"] <= q3)]

# 6Ô∏è‚É£ Test de Tukey
model = ols('Delivery_Time_min ~ C(Weather)', data=livraison).fit()
tukey = pairwise_tukeyhsd(endog=livraison['Delivery_Time_min'],
                          groups=livraison['Weather'],
                          alpha=0.05)

# 7Ô∏è‚É£ R√©sum√© Tukey
from statsmodels.graphics.factorplots import interaction_plot
tukey_summary = pd.DataFrame(data=tukey.summary().data[1:], columns=tukey.summary().data[0])

# 8Ô∏è‚É£ Visualisation
plt.figure(figsize=(10, 6))
tukey.plot_simultaneous(comparison_name='Clear', xlabel='Diff√©rence moyenne (min)', ylabel='M√©t√©o')
plt.title("Comparaison multiple : Test de Tukey HSD")
plt.grid(True)
plt.tight_layout()
plt.show()
```

<div style="border: 2px solid #dedede; border-radius: 10px; padding: 20px; background-color: #f9f9fc; font-family: 'Segoe UI', sans-serif; line-height: 1.6;">
  <h2 style="color:#4A90E2; margin-top: 0;">üìä Interpr√©tation du Test de Tukey HSD</h2>

  <p>Le graphique ci-dessus illustre les <strong>comparaisons multiples post-hoc</strong> entre les diff√©rentes conditions m√©t√©orologiques apr√®s une ANOVA significative.</p>

  <ul>
    <li><strong>Les barres horizontales</strong> repr√©sentent des intervalles de confiance autour de la diff√©rence moyenne de d√©lai de livraison entre deux types de m√©t√©o.</li>
    <li><span style="color:crimson; font-weight:600;">Si l'intervalle ne croise pas la ligne verticale (diff√©rence = 0)</span>, cela signifie que la diff√©rence est <strong>statistiquement significative</strong>.</li>
    <li><span style="color:gray;">Exemple :</span> La comparaison entre <code>Clear</code> et <code>Snowy</code> montre une diff√©rence significative car la barre ne croise pas 0.</li>
    <li>Les paires avec une barre <span style="color:gray;">qui croise la ligne verticale</span> ne pr√©sentent pas de diff√©rence significative.</li>
  </ul>

  <p style="margin-top: 15px;">
    ‚úÖ <strong>Conclusion :</strong> Certaines conditions m√©t√©o comme <strong>Foggy</strong>, <strong>Rainy</strong>, et <strong>Snowy</strong> allongent significativement le d√©lai de livraison compar√© √† des conditions <strong>Clear</strong> ou <strong>Windy</strong>.
  </p>

  <p style="background-color: #fff8e1; border-left: 4px solid #f5a623; padding: 10px; margin-top: 20px;">

  </p>
</div>


<div style="background-color:#f9f9f9; border: 2px solid #e1e1e1; border-radius: 10px; padding: 20px; font-family: 'Segoe UI', sans-serif;">
  <h2 style="color:#2c3e50;">üìà 7. Repr√©sentation graphique</h2>
  <ul style="line-height: 1.8; font-size: 16px; color:#333;">
    <li>üì¶ <strong>Boxplot group√©</strong> du <em>d√©lai de livraison</em> par type de m√©t√©o</li>
    <li>üìä <strong>Moyenne ¬± IC</strong> (Intervalle de Confiance) par condition m√©t√©orologique</li>
    <li>üîç <strong>Diagramme des comparaisons significatives</strong> √† partir du test de Tukey HSD</li>
  </ul>

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from statsmodels.stats.multicomp import MultiComparison
import warnings


# Cr√©ation des visualisations
plt.figure(figsize=(15, 12))

# 1. Boxplot group√©
plt.subplot(3, 1, 1)
sns.boxplot(x='Weather', y='Delivery_Time_min', data=livraison, palette="Set2")
plt.title("Boxplot du d√©lai de livraison par m√©t√©o")
plt.xlabel("Type de m√©t√©o")
plt.ylabel("D√©lai de livraison (minutes)")

# 2. Moyenne ¬± IC95%
plt.subplot(3, 1, 2)
group_stats = livraison.groupby('Weather')['Delivery_Time_min'].agg(['mean', 'count', 'std'])
group_stats['ci95'] = 1.96 * group_stats['std'] / group_stats['count']**0.5
group_stats = group_stats.sort_values(by='mean')
plt.bar(group_stats.index, group_stats['mean'], yerr=group_stats['ci95'], capsize=5, color='skyblue')
plt.title("Moyenne ¬± IC95% du d√©lai de livraison par m√©t√©o")
plt.ylabel("D√©lai moyen (minutes)")
plt.xlabel("Type de m√©t√©o")

# 3. Diagramme du test de Tukey
mc = MultiComparison(livraison['Delivery_Time_min'], livraison['Weather'])
tukey_result = mc.tukeyhsd()

plt.subplot(3, 1, 3)
tukey_result.plot_simultaneous(ylabel="M√©t√©o", xlabel="Diff√©rence moyenne (min)")
plt.title("Comparaison multiple : Test de Tukey HSD")

plt.tight_layout()
plt.show()
```
