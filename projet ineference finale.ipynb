Elles permettent de calculer des indicateurs statistiques (moyenne, m√©diane, √©cart-type) et sont souvent soumises √† des tests d‚Äôhypoth√®ses ou des mod√©lisations.

  <h3 style="color: #884ea0; margin-bottom: 10px;">üß† Pourquoi la m√©thode IQR ?</h3>

  <p style="font-size: 16px; color: #2c3e50; line-height: 1.6;">
    La m√©thode <strong>IQR (Interquartile Range)</strong> a √©t√© choisie pour d√©tecter les valeurs aberrantes car :
  </p>

  <ul style="font-size: 16px; color: #2c3e50; line-height: 1.6; list-style: none; padding-left: 0;">
    <li>‚úÖ Elle ne fait <strong>aucune hypoth√®se sur la distribution</strong> des donn√©es (contrairement au Z-score qui suppose une distribution normale).</li>
    <li>‚úÖ Elle est <strong>robuste aux valeurs extr√™mes</strong>, car elle se base sur les quartiles (Q1 et Q3).</li>
    <li>‚úÖ Elle est <strong>simple √† appliquer et √† interpr√©ter</strong>, tout en √©tant statistiquement fiable.</li>
  </ul>

  <p style="margin-top: 15px; font-size: 16px; color: #c0392b;">
    üéØ Dans un contexte de donn√©es r√©elles comme celui-ci (√©tudiants de profils vari√©s), la m√©thode IQR est id√©ale pour rep√©rer les cas atypiques sans fausser l‚Äôanalyse globale.
  </p>
</div>
lobale.


```python
# √âtape 1 : Nettoyage des noms de colonnes (au cas o√π)
student.columns = student.columns.str.strip()

# √âtape 2 : Fonction de d√©tection des valeurs aberrantes
def detect_outliers_iqr(df, seuil=1.5):
    outliers_info = []

    # On s√©lectionne uniquement les colonnes num√©riques
    num_cols = df.select_dtypes(include=['float64', 'int64']).columns

    for col in num_cols:
        Q1 = df[col].quantile(0.25)
        Q3 = df[col].quantile(0.75)
        IQR = Q3 - Q1
        borne_inf = Q1 - seuil * IQR
        borne_sup = Q3 + seuil * IQR

        # Identification des outliers
        outliers = df[(df[col] < borne_inf) | (df[col] > borne_sup)]
        nb_outliers = outliers.shape[0]
        pourcentage = (nb_outliers / df.shape[0]) * 100

        outliers_info.append({
            "Variable": col,
            "Nb_outliers": nb_outliers,
            "Pourcentage (%)": round(pourcentage, 2)
        })

    return pd.DataFrame(outliers_info)

# √âtape 3 : Application de la fonction sur le DataFrame
outliers_df = detect_outliers_iqr(student)

# √âtape 4 : Affichage du tableau r√©sultat
print("üîç Tableau des valeurs aberrantes (m√©thode IQR) :")
display(outliers_df)
```

# winsorization des valeurs extremes quantitative 

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats.mstats import winsorize
from IPython.display import display

# üì• Chargement des donn√©es

student.columns = student.columns.str.strip()

# üßÆ Variables quantitatives (sauf ID)
quantitative_vars = student.select_dtypes(include=['float64', 'int64']).columns.tolist()
quantitative_vars = [col for col in quantitative_vars if col != "id"]

# üîÅ Cr√©ation d'une copie pour comparaison avant traitement
student_before = student.copy()

# üß™ Application directe de la winsorisation √† 'student'
for col in quantitative_vars:
    student[col] = winsorize(student[col], limits=(0.01, 0.01))

# üìä Comparaison visuelle : AVANT vs APR√àS
plt.figure(figsize=(16, 12))

# üîπ AVANT winsorisation
plt.subplot(1, 2, 1)
sns.boxplot(data=student_before[quantitative_vars], orient='h', palette="coolwarm")
plt.title("Avant winsorisation", fontsize=14)
plt.grid(axis='x', linestyle='--', alpha=0.5)

# üîπ APR√àS winsorisation
plt.subplot(1, 2, 2)
sns.boxplot(data=student[quantitative_vars], orient='h', palette="coolwarm")
plt.title("Apr√®s winsorisation (modification directe)", fontsize=14)
plt.grid(axis='x', linestyle='--', alpha=0.5)

plt.tight_layout()
plt.show()

# ‚úÖ Aper√ßu des donn√©es finales
print("‚úÖ Donn√©es modifi√©es directement dans le DataFrame 'student' :")
display(student[quantitative_vars].head())

```

# üîπ Uniformisation de la dur√©e de sommeil : pr√©paration √† l‚Äôanalyse comportementale

```python
# üîß Fonction pour nettoyer et uniformiser la dur√©e de sommeil
def nettoyer_duree_sommeil(val):
    val = str(val).lower().strip()

    if "less" in val or val.startswith("4") or "<" in val or "moins" in val:
        return "< 5h"
    elif "5" in val and "6" in val:
        return "5-6h"
    elif "7" in val or "8" in val:
        return "7-8h"
    elif "more" in val or "9" in val or "10" in val or "plus" in val or ">" in val:
        return "> 8h"
    else:
        return "Non pr√©cis√©"

# üß™ Application DIRECTE sur la colonne d'origine
student["duree_sommeil"] = student["duree_sommeil"].apply(nettoyer_duree_sommeil)

# üìä V√©rification des r√©sultats
print(student["duree_sommeil"].value_counts())
```

```python
from scipy.stats import chi2_contingency, ttest_ind
# Nettoyage de la variable duree_sommeil
def nettoyer_duree_sommeil(val):
    val = str(val).lower().strip()
    if "less" in val or val.startswith("4") or "<" in val or "moins" in val:
        return "< 5h"
    elif "5" in val and "6" in val:
        return "5-6h"
    elif "7" in val or "8" in val:
        return "7-8h"
    elif "more" in val or "9" in val or "10" in val or "plus" in val or ">" in val:
        return "> 8h"
    else:
        return "Non pr√©cis√©"

student["duree_sommeil"] = student["duree_sommeil"].apply(nettoyer_duree_sommeil)

# Listes des variables
qual_vars = ['sexe', 'ville', 'profession', 'duree_sommeil', 
             'habitudes_alimentaires', 'diplome_suivi', 
             'pensees_suicidaire', 'antecedants_familiaux_maladie_mentale']

quant_vars = ['age', 'pression_academique', 'pression_liee_au_travail', 'moyenne_notes', 
              'satisfaction_etudes', 'satisfaction_travail', 
              'nombre_heure_travail_etude', 'stress_financier']

target = 'depression'

# Analyse bivari√©e automatis√©e
results = []

# üîπ Qualitatives : Chi¬≤ + barplots group√©s
for var in qual_vars:
    print(f"üîç Variable qualitative : {var}")
    cont_table = pd.crosstab(student[var], student[target])
    chi2, pval, _, _ = chi2_contingency(cont_table)
    results.append({'Variable': var, 'Test': 'Chi2', 'p-value': round(pval, 4)})
    
    # Visualisation
    plt.figure(figsize=(7, 4))
    sns.countplot(data=student, x=var, hue=target, palette='Set2')
    plt.title(f"R√©partition de {var} selon la d√©pression")
    plt.xticks(rotation=30)
    plt.tight_layout()
    plt.show()

# üî∏ Quantitatives : T-test + boxplots
for var in quant_vars:
    print(f"üî¨ Variable quantitative : {var}")
    g0 = student[student[target] == 0][var].dropna()
    g1 = student[student[target] == 1][var].dropna()
    stat, pval = ttest_ind(g0, g1)
    results.append({'Variable': var, 'Test': 'T-test', 'p-value': round(pval, 4)})
    
    # Visualisation
    plt.figure(figsize=(6, 4))
    sns.boxplot(x=target, y=var, data=student, palette='pastel')
    plt.title(f"{var} selon la d√©pression")
    plt.tight_layout()
    plt.show()

# üßæ Tableau des r√©sultats
results_df = pd.DataFrame(results)
print("üìä R√©sum√© des tests statistiques :")
print(results_df)
```

  <h3 style="color:#d35400;">üîπ Question 1 ‚Äî Intervalle de confiance : Pens√©es suicidaires</h3>

  <p><strong>üéØ Objectif :</strong> Estimer, avec une marge d‚Äôincertitude ma√Ætris√©e, la proportion r√©elle d‚Äô√©tudiants ayant d√©j√† eu des pens√©es suicidaires au sein de notre √©chantillon.</p>

  <p><strong>üìå Variable √©tudi√©e :</strong> <code>pensees_suicidaire</code></p>

  <p><strong>üß™ M√©thode utilis√©e :</strong> Intervalle de confiance √† 95 % pour une proportion, selon la m√©thode normale (approximation de Wald), fond√©e sur l‚Äôestimation de la proportion observ√©e dans un grand √©chantillon.</p>

  <p><strong>‚úÖ √âtapes m√©thodologiques :</strong></p>
  <ul style="line-height: 1.7;">
    <li>‚úîÔ∏è Calcul de la proportion observ√©e dans l‚Äô√©chantillon : <code>pÃÇ = x / n</code></li>
    <li>‚úîÔ∏è Calcul de l‚Äôerreur standard : <code>SE = ‚àö[pÃÇ(1 - pÃÇ) / n]</code></li>
    <li>‚úîÔ∏è Construction de l‚Äôintervalle de confiance √† 95 % : <code>[pÃÇ - 1.96 √ó SE ; pÃÇ + 1.96 √ó SE]</code></li>
  </ul>

  <p><strong>üìä Interpr√©tation :</strong> L‚Äôintervalle ainsi obtenu encadre, avec un niveau de confiance de 95 %, la proportion r√©elle d‚Äô√©tudiants susceptibles d‚Äôavoir connu ce type de souffrance mentale. Il permet une √©valuation robuste de la pr√©valence de cette probl√©matique dans la population √©tudiante.</p>

</div>


```python
import pandas as pd
import numpy as np
from scipy.stats import norm


# üéØ Cible : √âtudiants ayant d√©j√† eu des pens√©es suicidaires
# Nettoyage et transformation en binaire (Yes = 1, No = 0)
student['pensees_suicidaires_bin'] = student['pensees_suicidaire'].str.strip().str.lower().map({'yes': 1, 'no': 0})

# üîç Filtrage des valeurs valides
valid_data = student['pensees_suicidaires_bin'].dropna()

# üìä Param√®tres pour l'intervalle
n = valid_data.count()
x = valid_data.sum()
p_hat = x / n

# üîê Calcul de l'intervalle de confiance √† 95 %
z = norm.ppf(0.975)  # z-score pour 95 %
se = np.sqrt(p_hat * (1 - p_hat) / n)
lower = p_hat - z * se
upper = p_hat + z * se

# üìà R√©sum√© des r√©sultats
print("üìä Analyse : Intervalle de confiance √† 95 % pour la proportion d'√©tudiants ayant eu des pens√©es suicidaires")
print(f"Nombre total d'√©tudiants valides : {n}")
print(f"Nombre d'√©tudiants d√©clarant des pens√©es suicidaires : {x}")
print(f"Proportion estim√©e : {p_hat:.3f}")
print(f"‚úÖ Intervalle de confiance (95 %) : [{lower:.3f}, {upper:.3f}]")
```

```python
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm

# üß† Facultatif : D√©finir une police compatible emoji si disponible
plt.rcParams['font.family'] = 'DejaVu Sans'  # G√©n√©ralement compatible avec matplotlib
# plt.rcParams['font.family'] = 'Segoe UI Emoji'  # üí° Optionnel si tu as cette police sur ton syst√®me

# üî¢ Donn√©es d√©j√† calcul√©es
proportion = p_hat
lower_bound = lower
upper_bound = upper
margin_error = upper_bound - proportion

# üìä Cr√©ation du graphique
plt.figure(figsize=(7, 6))
plt.bar(['üß† Pens√©es suicidaires'], [proportion],
        color='#85C1E9', edgecolor='black', alpha=0.85,
        width=0.5, label=f"Proportion estim√©e : {proportion:.3f}")

# üß™ Ajout de l‚Äôintervalle de confiance
plt.errorbar(['üß† Pens√©es suicidaires'], [proportion],
             yerr=[[proportion - lower_bound], [upper_bound - proportion]],
             fmt='o', color='black', capsize=10, linewidth=2,
             label='Intervalle de confiance √† 95 %')

# ‚úèÔ∏è Annotation texte
plt.text(0, proportion + 0.02, f"{proportion*100:.1f} %",
         ha='center', va='bottom', fontsize=12, fontweight='bold', color='#154360')

# üé® Mise en forme
plt.title("üìä Proportion d'√©tudiants ayant eu des pens√©es suicidaires\navec intervalle de confiance √† 95 %",
          fontsize=13, fontweight='bold', color="#1b2631")
plt.ylim(0, 1)
plt.ylabel("Proportion", fontsize=12)
plt.xticks(fontsize=11)
plt.yticks(fontsize=11)
plt.grid(axis='y', linestyle='--', alpha=0.4)
plt.legend(loc='upper right')
plt.tight_layout()
plt.show()
```

  <h3 style="color: #2e86c1;">üìå Justification m√©thodologique : choix de la m√©thode Bootstrap</h3>

  <p>Dans le cadre de notre analyse des <strong>heures d‚Äô√©tudes/travail</strong> des √©tudiants souffrant de d√©pression, nous avons souhait√© estimer la <strong>moyenne</strong> et construire un <strong>intervalle de confiance</strong> √† 95 %.</p>

  <h4 style="margin-top: 15px;">üß™ √âtape 1 ‚Äì Test de normalit√©</h4>
  <p>Nous avons utilis√© le <strong>test de Shapiro-Wilk</strong> pour v√©rifier l‚Äôhypoth√®se de normalit√©. Le test a retourn√© une <strong>p-valeur extr√™mement faible</strong>, indiquant que <u>la distribution n‚Äôest pas normale</u>.</p>

  <h4 style="margin-top: 15px;">‚ö†Ô∏è Probl√®me</h4>
  <p>Les <strong>m√©thodes param√©triques classiques</strong> (bas√©es sur la loi de Student) supposent une normalit√© des donn√©es. Leur application dans ce contexte pourrait conduire √† des r√©sultats biais√©s.</p>

  <h4 style="margin-top: 15px;">üéØ Objectif</h4>
  <p>Fournir une estimation robuste de la moyenne, accompagn√©e d‚Äôun <strong>intervalle de confiance fiable</strong>, m√™me sans normalit√©.</p>

  <h4 style="margin-top: 15px;">‚úÖ Choix de la m√©thode : <span style="color: #117864;">Bootstrap</span></h4>
  <ul style="line-height: 1.7;">
    <li><strong>Ind√©pendante de toute loi de distribution</strong> (pas besoin de normalit√©)</li>
    <li><strong>Simple √† impl√©menter</strong> via resampling</li>
    <li><strong>Robuste</strong> m√™me en pr√©sence de donn√©es asym√©triques</li>
    <li>Id√©ale dans un cadre d‚Äôanalyse rigoureuse comme le n√¥tre</li>
  </ul>

  <p style="margin-top: 10px;">üîê <strong>Conclusion :</strong> la m√©thode <span style="color: #117864;">bootstrap</span> est adopt√©e pour estimer l‚Äôintervalle de confiance autour de la moyenne du nombre d‚Äôheures de travail chez les √©tudiants d√©pressifs.</p>
</div>


```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Filtrer les √©tudiants souffrant de d√©pression (depression = 1)
depressifs = student[student['depression'] == 1]['nombre_heure_travail_etude'].dropna()

# Fonction bootstrap pour la moyenne
def bootstrap_mean(data, n_iterations=1000, ci=95):
    np.random.seed(42)
    means = []
    for _ in range(n_iterations):
        sample = np.random.choice(data, size=len(data), replace=True)
        means.append(np.mean(sample))
    
    lower = np.percentile(means, (100 - ci) / 2)
    upper = np.percentile(means, 100 - (100 - ci) / 2)
    return np.mean(data), lower, upper, means

# Calcul bootstrap
mean_val, lower_ci, upper_ci, boot_dist = bootstrap_mean(depressifs)

# Affichage des r√©sultats
print("üìä Moyenne (bootstrap) :", round(mean_val, 2))
print(f"üìè Intervalle de confiance √† 95 % : [{round(lower_ci, 2)}, {round(upper_ci, 2)}]")

# Visualisation
plt.figure(figsize=(10, 5))
sns.histplot(boot_dist, kde=True, color="#3498db", bins=30)
plt.axvline(x=lower_ci, color='red', linestyle='--', label='Limite inf√©rieure')
plt.axvline(x=upper_ci, color='red', linestyle='--', label='Limite sup√©rieure')
plt.axvline(x=mean_val, color='green', linestyle='-', label='Moyenne')
plt.title("üìò Bootstrap : Moyenne des heures d'√©tude (√©tudiants d√©pressifs)", fontsize=14)
plt.xlabel("Moyenne simul√©e")
plt.ylabel("Fr√©quence")
plt.legend()
plt.grid(True, linestyle='--', alpha=0.4)
plt.tight_layout()
plt.show()
```

  <h3 style="color:#884ea0;">üìå Justification m√©thodologique : Estimation de la m√©diane par Bootstrap</h3>

  <p>
    L‚Äôobjectif est d‚Äôestimer la <strong>m√©diane</strong> du nombre d‚Äôheures de travail ou d‚Äô√©tudes chez les √©tudiants souffrant de d√©pression, avec un <strong>intervalle de confiance fiable</strong>.
  </p>

  <p>
    Contrairement √† la moyenne, il n‚Äôexiste pas de formule param√©trique simple pour construire un intervalle de confiance autour de la <strong>m√©diane</strong>, surtout lorsque les donn√©es sont <strong>non normales</strong>, comme c‚Äôest le cas ici.
  </p>

  <p>
    La m√©thode <strong>bootstrap</strong> s‚Äôimpose donc naturellement pour cette estimation. Elle repose sur un processus de <em>resampling</em> avec remise, qui permet d‚Äô√©valuer de mani√®re empirique la distribution de la m√©diane sans supposer de forme particuli√®re des donn√©es.
  </p>

  <ul style="line-height: 1.7;">
    <li>‚úÖ Ne d√©pend d‚Äôaucune hypoth√®se de normalit√©</li>
    <li>‚úÖ Robuste face aux distributions asym√©triques</li>
    <li>‚úÖ Fournit des intervalles cr√©dibles et adaptatifs</li>
  </ul>

  <p style="margin-top: 10px;">
    üîê <strong>Conclusion :</strong> le <strong>bootstrap</strong> est la m√©thode la plus fiable et la plus rationnelle pour estimer l‚Äôintervalle de confiance de la m√©diane dans notre contexte, garantissant la rigueur et la robustesse de notre analyse.
  </p>

</div>


```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# üéØ Extraction des valeurs non nulles pour les √©tudiants d√©pressifs
depressifs = student[student['depression'] == 1]['nombre_heure_travail_etude'].dropna()

# üîÅ Fonction bootstrap pour la m√©diane
def bootstrap_median(data, n_iterations=1000, ci=95):
    np.random.seed(42)
    medians = []
    for _ in range(n_iterations):
        sample = np.random.choice(data, size=len(data), replace=True)
        medians.append(np.median(sample))
    
    lower = np.percentile(medians, (100 - ci) / 2)
    upper = np.percentile(medians, 100 - (100 - ci) / 2)
    return np.median(data), lower, upper, medians

# üîç Estimation de la m√©diane et de l'intervalle de confiance
mediane_val, lower_ci, upper_ci, boot_dist = bootstrap_median(depressifs)

# ‚úÖ R√©sultats
print("üìä M√©diane (bootstrap) :", round(mediane_val, 2))
print(f"üìè Intervalle de confiance √† 95 % : [{round(lower_ci, 2)}, {round(upper_ci, 2)}]")

# üé® Visualisation de la distribution bootstrap
plt.figure(figsize=(10, 5))
sns.histplot(boot_dist, kde=True, color="#af7ac5", bins=30)
plt.axvline(x=lower_ci, color='red', linestyle='--', label='Limite inf√©rieure')
plt.axvline(x=upper_ci, color='red', linestyle='--', label='Limite sup√©rieure')
plt.axvline(x=mediane_val, color='green', linestyle='-', label='M√©diane observ√©e')
plt.title("Bootstrap : M√©diane des heures d'√©tude (√©tudiants d√©pressifs)", fontsize=14)
plt.xlabel("M√©diane simul√©e")
plt.ylabel("Fr√©quence")
plt.legend()
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

  <p>Pour comparer les <strong>heures de travail ou d‚Äô√©tudes</strong> entre les √©tudiants souffrant de <code>d√©pression</code> (<em>groupe 1</em>) et ceux ne souffrant pas (<em>groupe 0</em>), nous avons suivi les √©tapes m√©thodologiques suivantes :</p>

  <ul style="line-height: 1.8;">
    <li>üîç <strong>Objectif :</strong> D√©terminer si une diff√©rence significative existe entre les deux groupes sur le plan de la charge acad√©mique.</li>
    <li>üìâ <strong>V√©rification de la normalit√© :</strong> Le <em>test de Shapiro-Wilk</em> a √©t√© appliqu√© sur la variable <code>nombre_heure_travail_etude</code>. R√©sultat : <strong>la normalit√© est rejet√©e</strong> (p-valeur ‚â™ 0.05).</li>
    <li>üìä <strong>Conclusion :</strong> En l‚Äôabsence de normalit√©, le test param√©trique (t-test) n‚Äôest pas appropri√©.</li>
    <li>‚úÖ <strong>Test utilis√© :</strong> Le <strong>test de Mann‚ÄìWhitney</strong> (non param√©trique) a √©t√© choisi pour comparer les <strong>m√©dianes</strong> entre les deux groupes.</li>
    <li>üí° <strong>Hypoth√®se nulle (H‚ÇÄ) :</strong> Les distributions des heures de travail sont identiques pour les deux groupes.</li>
  </ul>

  <p style="margin-top: 12px;">Le test s‚Äôest av√©r√© significatif, justifiant une diff√©rence de comportement entre les √©tudiants selon leur √©tat d√©pressif.</p>

```python
import pandas as pd
from scipy.stats import mannwhitneyu
import matplotlib.pyplot as plt
import seaborn as sns

# üîç Cr√©ation des deux groupes
depressifs = student[student["depression"] == 1]["nombre_heure_travail_etude"].dropna()
non_depressifs = student[student["depression"] == 0]["nombre_heure_travail_etude"].dropna()

# üß™ Test de Mann‚ÄìWhitney (non param√©trique)
stat, p_value = mannwhitneyu(depressifs, non_depressifs, alternative='two-sided')

# ‚úÖ Affichage du r√©sultat
print("üìä Statistique de test Mann‚ÄìWhitney :", stat)
print("üìå P-valeur :", p_value)

# üé® Visualisation du r√©sultat par boxplot
plt.figure(figsize=(10, 6))
sns.boxplot(x="depression", y="nombre_heure_travail_etude", data=student, palette=["#f5b7b1", "#aed6f1"])
plt.title("Comparaison des heures d‚Äô√©tude selon l‚Äô√©tat d√©pressif", fontsize=14, fontweight='bold')
plt.xlabel("√âtat d√©pressif (0 = Non, 1 = Oui)", fontsize=12)
plt.ylabel("Heures de travail / √©tude", fontsize=12)
plt.grid(axis="y", linestyle="--", alpha=0.6)
plt.tight_layout()
plt.show()
```

  <h3 style="color: #2e86c1;">üìä R√©sultat du test de Mann‚ÄìWhitney</h3>

  <p>Le test de <strong>Mann‚ÄìWhitney</strong> a √©t√© r√©alis√© pour comparer les <strong>heures de travail ou d‚Äô√©tudes</strong> entre les √©tudiants souffrant de d√©pression (<code>depression = 1</code>) et ceux qui n‚Äôen souffrent pas (<code>depression = 0</code>).</p>

Cela signifie qu‚Äôil existe <strong>une diff√©rence significative</strong> entre les deux groupes quant au nombre d‚Äôheures de travail ou d‚Äô√©tude.

  <p style="margin-top: 8px;">Ce r√©sultat sugg√®re que la charge acad√©mique pourrait √™tre un facteur diff√©renciateur chez les √©tudiants en situation de d√©tresse psychologique. Il conviendrait d‚Äôexplorer cette relation plus en profondeur dans la suite de l‚Äôanalyse.</p>

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt


# üîÅ Fonction de bootstrap pour moyenne ou m√©diane
def bootstrap_ci(data, stat_func=np.mean, n_bootstrap=10000, ci=95):
    np.random.seed(42)
    boot_stats = [stat_func(np.random.choice(data, size=len(data), replace=True)) for _ in range(n_bootstrap)]
    lower = np.percentile(boot_stats, (100 - ci) / 2)
    upper = np.percentile(boot_stats, 100 - (100 - ci) / 2)
    return np.mean(boot_stats), (lower, upper), boot_stats

# üßπ Suppression des valeurs manquantes
group_depressif = student[student["depression"] == 1]["stress_financier"].dropna()
group_non_depressif = student[student["depression"] == 0]["stress_financier"].dropna()

# üìä Bootstrap sur les deux groupes
mean_dep, ci_mean_dep, boot_mean_dep = bootstrap_ci(group_depressif, np.mean)
med_dep, ci_med_dep, boot_med_dep = bootstrap_ci(group_depressif, np.median)

mean_nondep, ci_mean_nondep, boot_mean_nondep = bootstrap_ci(group_non_depressif, np.mean)
med_nondep, ci_med_nondep, boot_med_nondep = bootstrap_ci(group_non_depressif, np.median)

# ‚úÖ R√©sultats
print("GROUPE : √âTUDIANTS D√âPRESSIFS")
print(f"Moyenne : {mean_dep:.3f} | IC95% : [{ci_mean_dep[0]:.2f}, {ci_mean_dep[1]:.2f}]")
print(f"M√©diane : {med_dep:.1f}   | IC95% : [{ci_med_dep[0]:.1f}, {ci_med_dep[1]:.1f}]\n")

print("GROUPE : √âTUDIANTS NON D√âPRESSIFS")
print(f"Moyenne : {mean_nondep:.3f} | IC95% : [{ci_mean_nondep[0]:.2f}, {ci_mean_nondep[1]:.2f}]")
print(f"M√©diane : {med_nondep:.1f}   | IC95% : [{ci_med_nondep[0]:.1f}, {ci_med_nondep[1]:.1f}]")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import shapiro, mannwhitneyu, ttest_ind

# üßπ Suppression des valeurs manquantes
student = student.dropna(subset=["satisfaction_etudes", "depression"])

# üß™ S√©paration des groupes
group_dep = student[student["depression"] == 1]["satisfaction_etudes"]
group_non_dep = student[student["depression"] == 0]["satisfaction_etudes"]

# üìä V√©rification de la normalit√© (Shapiro-Wilk)
shapiro_dep = shapiro(group_dep.sample(n=5000, random_state=1) if len(group_dep) > 5000 else group_dep)
shapiro_non_dep = shapiro(group_non_dep.sample(n=5000, random_state=1) if len(group_non_dep) > 5000 else group_non_dep)

# üîç Choix du test selon la normalit√©
if shapiro_dep.pvalue < 0.05 or shapiro_non_dep.pvalue < 0.05:
    test_name = "Mann‚ÄìWhitney"
    stat, pval = mannwhitneyu(group_dep, group_non_dep)
else:
    test_name = "T-test"
    stat, pval = ttest_ind(group_dep, group_non_dep, equal_var=False)

# üìà R√©sum√© des r√©sultats
resultats = {
    "Test utilis√©": test_name,
    "Statistique": round(stat, 4),
    "P-valeur": round(pval, 6),
    "Moyenne (d√©pressifs)": round(group_dep.mean(), 2),
    "Moyenne (non d√©pressifs)": round(group_non_dep.mean(), 2),
    "M√©diane (d√©pressifs)": round(group_dep.median(), 2),
    "M√©diane (non d√©pressifs)": round(group_non_dep.median(), 2)
}

# üé® Visualisation par boxplot
plt.figure(figsize=(9, 6))
sns.boxplot(x="depression", y="satisfaction_etudes", data=student, palette=["#f1948a", "#85c1e9"])
plt.title("Satisfaction des √©tudes selon l‚Äô√©tat d√©pressif", fontsize=14, fontweight='bold')
plt.xlabel("D√©pression (0 = Non, 1 = Oui)")
plt.ylabel("Satisfaction des √©tudes")
plt.grid(axis="y", linestyle="--", alpha=0.5)
plt.tight_layout()
plt.show()

# üñ®Ô∏è Affichage du dictionnaire de r√©sultats
print(resultats)
```

    <li>‚öñÔ∏è Application du <strong>test de Kruskal-Wallis</strong> (non param√©trique) pour tester l‚Äôhypoth√®se d‚Äôune diff√©rence de satisfaction entre dipl√¥mes.</li>
  </ul>

  <h3 style="color:#34495e;">‚ö†Ô∏è R√©sultat :</h3>
  <p>
    Le test n‚Äôa pas pu √™tre ex√©cut√© car <span style="color:#e74c3c; font-weight:bold;">les valeurs de satisfaction au travail √©taient identiques dans tous les groupes</span>.
    Cela a g√©n√©r√© une erreur dans le test de Kruskal‚ÄìWallis (<em>¬´¬†All numbers are identical in kruskal¬†¬ª</em>).
  </p>

  <h3 style="color:#34495e;">üîç Limites et biais d√©tect√©s :</h3>
  <ul style="line-height:1.7;">
    <li>üìâ <strong>Manque de variabilit√© :</strong> La variable <code>satisfaction_travail</code> pr√©sente une distribution extr√™mement uniforme, limitant toute inf√©rence statistique fiable.</li>
    <li>üîÑ <strong>Possibles biais de mesure :</strong> Cela pourrait √™tre d√ª √† un biais de formulation du questionnaire, √† une tendance de r√©ponses sociales ou √† une √©chelle trop restreinte.</li>
    <li>üß© <strong>Discrimination faible :</strong> Cette variable semble peu discriminante pour diff√©rencier les profils selon le dipl√¥me.</li>
  </ul>

  <h3 style="color:#34495e;">üôè Remarque importante :</h3>
  <p style="color:#555;">
    Nous nous excusons pour l‚Äôimpossibilit√© d‚Äôaboutir √† une analyse robuste sur cette variable. Malgr√© un protocole rigoureux, 
    certaines variables peuvent pr√©senter des limitations intrins√®ques. Cette situation souligne l‚Äôimportance d‚Äôune 
    <strong>qualit√© de mesure</strong> ad√©quate dans toute enqu√™te statistique.
  </p>

  <p style="margin-top:10px;">
    üöÄ <strong>Suggestion :</strong> Pour approfondir, il serait int√©ressant d‚Äôutiliser une variable plus sensible (comme <code>satisfaction_etudes</code> ou <code>stress_financier</code>), 
    ou de repenser l‚Äô√©chelle de satisfaction pour les futurs sondages.
  </p>
</div>


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import kruskal

# 1. Suppression des valeurs manquantes
student = student.dropna(subset=["satisfaction_travail", "diplome_suivi"])

# 2. Regrouper les dipl√¥mes rares
counts = student["diplome_suivi"].value_counts()
valid_diplomes = counts[counts > 30].index
student["diplome_suivi_grouped"] = student["diplome_suivi"].apply(lambda x: x if x in valid_diplomes else "Autre")

# 3. Regrouper les donn√©es pour le test (seulement si les groupes ont une variance non nulle)
grouped_data = []
for name, group in student.groupby("diplome_suivi_grouped"):
    values = group["satisfaction_travail"].values
    if len(set(values)) > 1:  # V√©rifie que toutes les valeurs ne sont pas identiques
        grouped_data.append(values)

# 4. Test de Kruskal-Wallis
if len(grouped_data) >= 2:
    stat, p_value = kruskal(*grouped_data)
    print("‚úÖ Test de Kruskal‚ÄìWallis ‚Äì Satisfaction au travail")
    print(f"Statistique H : {stat:.2f}")
    print(f"P-valeur      : {p_value:.6e}")
else:
    print("‚ö†Ô∏è Pas assez de groupes valides (avec variabilit√©) pour ex√©cuter le test.")

# 5. Visualisation
plt.figure(figsize=(12, 6))
sns.boxplot(x="diplome_suivi_grouped", y="satisfaction_travail", data=student, palette="Set2")
plt.title("üìä Satisfaction au travail selon le dipl√¥me suivi", fontsize=14, fontweight='bold')
plt.xlabel("Dipl√¥me suivi")
plt.ylabel("Satisfaction au travail")
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.4)
plt.tight_layout()
plt.show()
```

  <h2 style="color: #8e44ad;">‚ùì Question d‚Äôanalyse :</h2>

  <p style="font-size: 16px; line-height: 1.6;">
    <strong>La d√©pression est-elle ind√©pendante des habitudes alimentaires</strong> (saines, mod√©r√©es ou malsaines) chez les √©tudiants ? Autrement dit :
  </p>

  <ul style="line-height: 1.8;">
    <li>üëâ Les comportements alimentaires influencent-ils l‚Äô√©tat de sant√© mentale ?</li>
    <li>üëâ Ou bien, les deux variables sont-elles statistiquement <strong>ind√©pendantes</strong> ?</li>
  </ul>

  <p style="margin-top: 10px;">
    Cette question appelle une <strong>analyse d‚Äôassociation</strong> entre deux variables qualitatives, qui sera r√©solue √† l‚Äôaide du <strong>test du Khi-deux</strong>.
  </p>

</div>


<div style="font-family: 'Segoe UI', sans-serif; background-color: #fefefe; border-left: 5px solid #2980b9; padding: 20px; border-radius: 8px; color: #333; box-shadow: 0 2px 5px rgba(0,0,0,0.04);">

  <h3 style="color: #2980b9; margin-bottom: 10px;">üß™ Justification m√©thodologique ‚Äì Choix du test du Chi¬≤</h3>

  <p>
    L‚Äôobjectif est de d√©terminer s‚Äôil existe une <strong>relation statistiquement significative</strong> entre deux variables qualitatives :
    <strong>la d√©pression</strong> (Oui / Non) et <strong>les habitudes alimentaires</strong> (Healthy / Moderate / Unhealthy / Others).
  </p>

  <p>
    ‚úÖ Pour ce faire, nous avons retenu le <strong>test d‚Äôind√©pendance du Khi-deux (Chi¬≤)</strong>, en raison des √©l√©ments suivants :
  </p>

  <ul style="line-height: 1.7;">
    <li>üî∏ Les deux variables sont <strong>cat√©gorielles</strong> (nominales ou ordinales).</li>
    <li>üî∏ L‚Äô√©chantillon est suffisamment <strong>grand</strong> pour garantir la validit√© des fr√©quences attendues.</li>
    <li>üî∏ Le test permet de comparer la <strong>distribution observ√©e</strong> √† une distribution th√©orique sous hypoth√®se d‚Äôind√©pendance.</li>
  </ul>

  <p>
    üìå Le test repose sur la comparaison des effectifs observ√©s et attendus dans un <strong>tableau de contingence</strong>.
    Si la p-valeur obtenue est inf√©rieure √† 5 %, on rejette l‚Äôhypoth√®se d‚Äôind√©pendance.
  </p>

  <p style="margin-top: 10px;">
    üëâ Le test du Chi¬≤ est donc parfaitement adapt√© pour <strong>√©valuer l'association entre l‚Äô√©tat mental et les comportements alimentaires</strong> 
    dans cette √©tude.
  </p>

</div>


```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import chi2_contingency

# 1. Table de contingence
contingency_table = pd.crosstab(student['depression'], student['habitudes_alimentaires'])

# 2. Test du Chi¬≤
chi2_stat, p_value, dof, expected = chi2_contingency(contingency_table)

# 3. R√©sultats du test
print("üìä Test d'ind√©pendance (Chi¬≤) entre la d√©pression et les habitudes alimentaires")
print(f"Statistique Chi¬≤ : {chi2_stat:.2f}")
print(f"P-valeur          : {p_value:.5f}")
print(f"Degr√©s de libert√© : {dof}")
print("\n‚úÖ Table de contingence observ√©e :")
print(contingency_table)

# 4. Visualisation (heatmap)
plt.figure(figsize=(8, 6))
sns.heatmap(contingency_table, annot=True, fmt="d", cmap="YlGnBu", cbar=False)
plt.title("üß† Distribution crois√©e : D√©pression vs Habitudes alimentaires", fontsize=13, weight='bold')
plt.xlabel("Habitudes alimentaires")
plt.ylabel("D√©pression")
plt.tight_layout()
plt.show()
```

  <h2 style="color: #2980b9;">üß≠ Objectif de l‚Äôanalyse</h2>

  <p style="font-size: 16px; line-height: 1.6;">
    L‚Äôobjectif de cette analyse est de <strong>v√©rifier l‚Äôexistence d‚Äôune relation statistiquement significative</strong> entre :
  </p>

  <ul style="font-size: 16px; line-height: 1.8;">
    <li><strong>Variable 1 :</strong> <span style="color: #34495e;">√âtat d√©pressif</span> (Oui / Non)</li>
    <li><strong>Variable 2 :</strong> <span style="color: #34495e;">Dur√©e du sommeil</span> (cat√©gories : <code>&lt; 5h</code>, <code>5‚Äì6h</code>, <code>7‚Äì8h</code>, <code>&gt; 8h</code>)</li>
  </ul>

  <p style="font-size: 16px;">
    Cette v√©rification sera effectu√©e √† l‚Äôaide d‚Äôun <strong>test d‚Äôind√©pendance du Khi-deux</strong>, afin de d√©terminer si la distribution du sommeil influence significativement la pr√©valence de la d√©pression.
  </p>

</div>


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import chi2_contingency

# üßΩ Fonction de nettoyage des dur√©es de sommeil
def nettoyer_duree(val):
    val = str(val).lower().strip()
    if "moins" in val or "<" in val or "4" in val:
        return "< 5h"
    elif "5" in val and "6" in val:
        return "5-6h"
    elif "7" in val or "8" in val:
        return "7-8h"
    elif "9" in val or "10" in val or "plus" in val or ">" in val:
        return "> 8h"
    else:
        return "Autre"

# Application de la fonction
student["duree_sommeil_clean"] = student["duree_sommeil"].apply(nettoyer_duree)

# Table de contingence
contingency_table = pd.crosstab(student["duree_sommeil_clean"], student["depression"])

# ‚úÖ Test du Khi-2
chi2_stat, p_value, dof, expected = chi2_contingency(contingency_table)

# Affichage du r√©sultat
print("üìä Test d‚Äôind√©pendance (Chi¬≤) : D√©pression vs Dur√©e de sommeil")
print(f"Statistique Chi¬≤ : {chi2_stat:.2f}")
print(f"P-valeur         : {p_value:.6f}")
print(f"Degr√©s de libert√© : {dof}")

# üìä Heatmap de visualisation
plt.figure(figsize=(8, 6))
sns.heatmap(contingency_table, annot=True, fmt="d", cmap="YlGnBu", cbar=False)
plt.title("Distribution crois√©e : D√©pression vs Dur√©e de sommeil", fontsize=14, fontweight='bold')
plt.xlabel("D√©pression")
plt.ylabel("Dur√©e de sommeil")
plt.tight_layout()
plt.show()
```
